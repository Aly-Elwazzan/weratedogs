# Assess tweet_json

```python



import json

with open('tweet_json2.txt') as f:
    
    data = json.load(f)
    
temp = []

for tweet_id in data.keys():
    retweets = data[tweet_id]['retweet_count']
    favorites = data[tweet_id]['favorite_count']
    temp.append({'tweet_id': tweet_id,
                        'retweets': retweets,
                        'favorites': favorites})
    
    
retweet_df = pd.DataFrame(temp,  columns=['tweet_id', 'retweets', 'favorites'])
```

# cleaning


```python
arch_clean=df_arch.copy()
retweet_clean=retweet_df.copy()
image_clean=df_image.copy()

```

### Define: dog stages in multiple columns(doggo, floofer, puppo, pupper)

### code:Melt dog stage columns into single column.

```python
arch_clean = pd.melt(arch_clean,
                        id_vars=['tweet_id', 'in_reply_to_status_id', 'in_reply_to_user_id',
                                 'timestamp', 'source', 'text', 'retweeted_status_id',
                                 'retweeted_status_user_id', 'retweeted_status_timestamp',
                                 'expanded_urls', 'rating_numerator', 'rating_denominator',
                                 'name',],
                        value_name='dog_stage')
```

### TEST


```python


arch_clean.info()
```

```python
arch_clean = arch_clean.drop('variable', axis=1)
```

```python
arch_clean.info()
```

```python
## remove duplicates now every thing is present 4 times
```

```python
dup_arch = arch_clean.drop_duplicates(subset=['tweet_id', 'dog_stage']) 
dup_arch.info()
 ## note there is still some duplicates 
```

```python

dup_pos = arch_clean.duplicated(subset='tweet_id', keep = False)  #remove all repeated ids
dup_df2 = arch_clean[dup_pos]
dup_df2 = dup_df2[dup_df2.dog_stage != 'None']

dup_df2.info()
```

```python
dup_df2.dog_stage.value_counts()
# there should be no "none"
```

```python
# Combine
dup_arch = dup_arch.drop_duplicates(subset = 'tweet_id', keep = False)
dup_df_merged = dup_arch.append(dup_df2)
```

```python
dup_df_merged.info()
```

```python
## merge 3 data sets, but all  tweet_id  must be same type
```

```python
dup_df_merged.tweet_id = dup_df_merged.tweet_id.astype(str)
image_clean.tweet_id = image_clean.tweet_id.astype(str)
```

### Define: three separate tables
### code : Convert tweet_id in archive_df to string. Inner join tweets_df, archive_df, and img_predictions on tweet_id to new column called dup_df_merged_new

```python
dup_df_merged_new = pd.merge(dup_df_merged, image_clean,
                        how = 'inner', on = 'tweet_id')


dup_df_merged_new= pd.merge(dup_df_merged_new, retweet_clean,
                        how = 'inner', on = 'tweet_id')
```

```python
dup_df_merged_new.info()
```

```python
# now lets fix  wrong  datatypes  
##(timestamp, retweeted_status_timestamp) to datetime
##(retweeted_status_id,retweeted_status_user_id)  to str
```

```python
dup_df_merged_new.timestamp =pd.to_datetime(dup_df_merged_new.timestamp)
dup_df_merged_new.retweeted_status_timestamp=pd.to_datetime(dup_df_merged_new.retweeted_status_timestamp)
```

```python
dup_df_merged_new.retweeted_status_id = dup_df_merged_new.retweeted_status_id.astype(str)
dup_df_merged_new.retweeted_status_user_id = dup_df_merged_new.retweeted_status_user_id.astype(str)
```
